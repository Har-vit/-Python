```python
def Division(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return 'Zero Division Error'
    except ValueError:
        return "Wrong input. It's must be numbers"
    
def Division_use():
    print(Division((int(input('Enter first number: '))), (int(input('Enter second number: ')))))
    print('\n')
 
def Data(**kwargs):
    return list(kwargs.values())
 
def Data_use():
    print(Data(name = input('Enter name: '),
                s_name = input('Enter second name: '),
                b_date = input('Enter birth day: '),
                l_town = input('Enter live town: '),
                email = input('Enter email: '),
                tel = input('Enter tel number: ')))
    print('\n')
 
def My_func(a, b, c):
    s = [a, b, c]
    s.remove(min(a, b, c))
    return sum(s)
 
 
def My_func_use():
    print(My_func((int(input('Enter first number: '))), (int(input('Enter second number: '))), (int(input('Enter third number: ')))))
    print('\n')
    
def My_func_2(x, y):
    return 1 / x ** abs(y)

def My_func_2_use():
    print(My_func_2((int(input('Enter first number: '))), (int(input('Enter second number: ')))))
    print('\n')
 
def My_func_2_1(x, y):
    for i in range(abs(y - 1)):
        x *= x
    return 1 / x

def My_func_2_1_use():
    print(My_func_2_1((int(input('Enter first number: '))), (int(input('Enter second number: ')))))
    print('\n')

def E_5():
    res = 0
    while True:
        numbers = input('Enter list of number or * to exit: ').split()
        for i in numbers:
            try:
                if i == '*':
                    print(f'Sum is {res}. Exit + "\n"')
                    return
                else:
                    res += int(i)
            except ValueError:
                print('Enter number or *')
        print(f'Sum is {res} + "\n"')

def Int_func(text):
    ls = []
    for i in range(len(text)):
        ls.append(text[i][0:1].title() + text[i][1:])
    return ' '.join(ls)
    print('\n')

def Int_func_use():
    print(Int_func(input('Input text: ').split()))
 
def functions():
    print('Задание #1. Введите: 1')
    print('Задание #2. Введите: 2')
    print('Задание #3. Введите: 3')
    print('Задание #4. Введите: 4 или 42')
    print('Задание #5. Введите: 5')
    print('Задание #6. Введите: 6' + '\n')
 
menu = {1: Division_use,
        2: Data_use,
        3: My_func_use,
        4: My_func_2_use,
        42: My_func_2_1_use,
        5: E_5,
        6: Int_func_use}
 
action = 1
 
while action != 0:
    functions()
    action = int(input('Выберите задание или введите "0", чтобы выйти: '))
    if action == 0:
        break
    if action in menu:
        x = menu[action]
        x()
```
