```python
def salary():
    x = int(input('Введите кол-во отработанных часов: '))
    y = int(input('Введите часовую ставку: '))
    c = int(input('Введите размер премии: '))
    pay = x * y
    return pay + c
print(f'з/п составила: {salary() }')
```

```python
list_1 = [4, -3, 6, -4, -11, 18, 13, 16, 9, 12, -3, 24, 99, 21]
list_2 = [list_1[i] for i in range(1, len(list_1)) if int(list_1[i]) > int(list_1[i - 1])]

print(list_2)
```

```python
from functools import reduce

def my_func(i1, i2):
    return i1 * i2

print(f'Список четных чисел {[i for i in range(99, 1001) if i % 2 == 0]}')
print(f'Результат перемножения всех элементов {reduce(my_func, [i for i in range(99, 1001) if i % 2 == 0])}')
```

```python
from itertools import count

for i in count(int(input('Введите число '))):
    print(i) 

from itertools import cycle

list_1 = [True, 'ABC', 123, None]
for i in cycle(list_1):
    print(i) 
```

```python
from itertools import count
from math import factorial

def fibo_gen():
    for i in count(1):
        yield factorial(i)

gen = fibo_gen()
x = 0
for i in gen:
    if x < 15:
        print(i)
        x += 1
    else:
        break 
```
